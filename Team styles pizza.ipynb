```python
import pandas as pd
from urllib.request import urlopen

import matplotlib.pyplot as plt
from PIL import Image

from mplsoccer import PyPizza, add_image, FontManager
```

```python
#standard stats
dfs = pd.read_html('https://fbref.com/en/comps/Big5/stats/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dfs):
 #print("***************************")
 #print(idx)
 #print(table)
stand = pd.DataFrame(dfs[0])
opp_stand = pd.DataFrame(dfs[1])
stand.drop(columns=["Per 90 Minutes"], inplace=True)
stand.columns = stand.columns.droplevel()
for_stand = stand[['Squad','Poss','npxG']]
opp_stand.drop(columns=["Per 90 Minutes"], inplace=True)
opp_stand.columns = opp_stand.columns.droplevel()
ag_stand = opp_stand[['Squad','npxG']]
final_stand = pd.merge(for_stand, ag_stand, left_index=True, right_index=True, suffixes=('', '_opp'))
final_stand.drop(columns=["Squad_opp"], inplace=True)
final_stand.head()
```

```python
#passing stats
dfp = pd.read_html('https://fbref.com/en/comps/Big5/passing/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dfp):
 #print("***************************")
 #print(idx)
 #print(table)
pas = pd.DataFrame(dfp[0])
opp_pas = pd.DataFrame(dfp[1])
pas.columns = pas.columns.droplevel()
cols = []
count = 1
for column in pas.columns:
    if column == 'Att':
        cols.append(f'Att_{count}')
        count+=1
        continue
    cols.append(column)
pas.columns = cols
for_pas = pas[['Squad','Att_1','TotDist','PrgDist']]
opp_pas.columns = opp_pas.columns.droplevel()
cols = []
count = 1
for column in opp_pas.columns:
    if column == 'Att':
        cols.append(f'Att_{count}')
        count+=1
        continue
    cols.append(column)
opp_pas.columns = cols
ag_pas = opp_pas[['Squad','Att_1']]
final_pas = pd.merge(for_pas, ag_pas, left_index=True, right_index=True, suffixes=('', '_opp'))
final_pas.drop(columns=["Squad_opp"], inplace=True)
stanpas = pd.merge(final_stand, final_pas, left_index=True, right_index=True, suffixes=('', '_opp'))
stanpas.drop(columns=["Squad_opp"], inplace=True)
stanpas.head()
```

```python
#GK stats
dfgk = pd.read_html('https://fbref.com/en/comps/Big5/keepersadv/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dfs):
 #print("***************************")
 #print(idx)
 #print(table)
keep = pd.DataFrame(dfgk[0])
keep.columns = keep.columns.droplevel()
cols = []
count = 1
for column in keep.columns:
    if column == 'Launch%':
        cols.append(f'Launch%_{count}')
        count+=1
        continue
    cols.append(column)
keep.columns = cols
keep.head()
keeper = keep[['Squad','Launch%_1']]
stanpasca = pd.merge(stanpas, keeper, left_index=True, right_index=True, suffixes=('', '_opp'))
stanpasca.drop(columns=["Squad_opp"], inplace=True)
stanpasca.head()
```

```python
#defensive stats
dft = pd.read_html('https://fbref.com/en/comps/Big5/defense/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dft):
 #print("***************************")
 #print(idx)
 #print(table)
tak = pd.DataFrame(dft[0])
opp_tak = pd.DataFrame(dft[1])
tak.columns = tak.columns.droplevel()
for_tak = tak[['Squad','Def 3rd','Mid 3rd','Att 3rd','Att','Tkl+Int']]
opp_tak.columns = opp_tak.columns.droplevel()
ag_tak = opp_tak[['Squad','Att','Tkl+Int']]
final_tak = pd.merge(for_tak, ag_tak, left_index=True, right_index=True, suffixes=('', '_opp'))
final_tak.drop(columns=["Squad_opp"], inplace=True)
stanpascat = pd.merge(stanpasca, final_tak, left_index=True, right_index=True, suffixes=('', '_opp'))
stanpascat.drop(columns=["Squad_opp"], inplace=True)
stanpascat.head()
```

```python
#possession stats
dftpo = pd.read_html('https://fbref.com/en/comps/Big5/possession/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dftpo):
 #print("***************************")
 #print(idx)
 #print(table)
pos = pd.DataFrame(dftpo[0])
opp_pos = pd.DataFrame(dftpo[1])
pos.columns = pos.columns.droplevel()
for_pos = pos[['Squad','Touches','Att 3rd']]
opp_pos.columns = opp_pos.columns.droplevel()
ag_pos = opp_pos[['Squad','Att 3rd']]
final_pos = pd.merge(for_pos, ag_pos, left_index=True, right_index=True, suffixes=('', '_opp'))
final_pos.drop(columns=["Squad_opp"], inplace=True)
stanpascatpo = pd.merge(stanpascat, final_pos, left_index=True, right_index=True, suffixes=('', '_opp'))
stanpascatpo.drop(columns=["Squad_opp"], inplace=True)
cols = []
count = 1
for column in stanpascatpo.columns:
    if column == 'Att 3rd_opp':
        cols.append(f'Att 3rd_opp_{count}')
        count+=1
        continue
    cols.append(column)
stanpascatpo.columns = cols
stanpascatpo = stanpascatpo.rename(columns={'Att 3rd_opp_1': 'Att 3rd touch', 'Att 3rd_opp_2': 'Att 3rd touch_opp'})
stanpascatpo.head()
```

```python
#pass type stats
dfcr = pd.read_html('https://fbref.com/en/comps/Big5/passing_types/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dfs):
 #print("***************************")
 #print(idx)
 #print(table)
cross = pd.DataFrame(dfcr[0])
cross.columns = cross.columns.droplevel()
crossing = cross[['Squad','Crs']]
bigfive = pd.merge(stanpascatpo, crossing, left_index=True, right_index=True, suffixes=('', '_opp'))
bigfive.drop(columns=["Squad_opp"], inplace=True)
bigfive.head()
```

```python
#shooting stats
dfsh = pd.read_html('https://fbref.com/en/comps/Big5/shooting/squads/Big-5-European-Leagues-Stats')
#for idx,table in enumerate(dfs):
 #print("***************************")
 #print(idx)
 #print(table)
shot = pd.DataFrame(dfsh[0])
shot.columns = shot.columns.droplevel()
shoot = shot[['Squad','npxG/Sh','Sh']]
bf = pd.merge(bigfive, shoot, left_index=True, right_index=True, suffixes=('', '_opp'))
bf.drop(columns=["Squad_opp"], inplace=True)
bf.head()
```

```python
bf.columns.values
```

```python
#setup metrics
team_style = bf[['Squad','Poss','Def 3rd','Mid 3rd','Att 3rd']]
team_style['Possession'] = bf['Poss']
team_style['Press resistance'] = bf['Att_1']/(bf['Tkl+Int_opp']+bf['Att_opp'])
team_style['Deep buildup'] = bf['Launch%_1']
team_style['Press intensity'] = bf['Att_1_opp']/(bf['Tkl+Int']+bf['Att'])
team_style['Chance prevention'] = bf['npxG_opp']
team_style['Def 3rd tackles'] = team_style['Def 3rd'] / (team_style['Def 3rd'] + team_style['Mid 3rd'] + team_style['Att 3rd']) 
team_style['Mid 3rd tackles'] = team_style['Mid 3rd'] / (team_style['Def 3rd'] + team_style['Mid 3rd'] + team_style['Att 3rd']) 
team_style['Att 3rd tackles'] = team_style['Att 3rd'] / (team_style['Def 3rd'] + team_style['Mid 3rd'] + team_style['Att 3rd']) 
team_style['Defensive height'] = (team_style['Def 3rd tackles']*16.5*1.1) + (team_style['Mid 3rd tackles']*49.5*1.1) + (team_style['Att 3rd tackles']*82.5*1.1)
team_style['Central progression'] = bf['Crs']/bf['Att_1']*100
team_style['Circulate'] = bf['PrgDist']*100/bf['TotDist']
team_style['Field tilt'] = bf['Att 3rd touch']/(bf['Att 3rd touch']+bf['Att 3rd touch_opp'])*100
team_style['Chance creation'] = bf['npxG']
team_style['Patient attack'] = bf['Sh']/bf['Att 3rd touch']*100
team_style['Shot quality'] = bf['npxG/Sh']
team_style.head()
```

```python
styles = team_style[['Squad','Chance prevention','Press intensity','Defensive height','Deep buildup','Press resistance','Possession','Central progression','Circulate','Field tilt','Chance creation','Patient attack','Shot quality']]
styles.head()
```

```python
styles['poss_rank'] = styles['Possession'].rank(pct=True)
styles['High line'] = styles['Defensive height'].rank(pct=True)
styles['Resistance rank'] = styles['Press resistance'].rank(pct=True)
styles['Tilt rank'] = styles['Field tilt'].rank(pct=True)
styles['Potency'] = styles['Chance creation'].rank(pct=True)
styles['Shooting'] = styles['Shot quality'].rank(pct=True)
styles['Tight def'] = styles['Chance prevention'].rank(pct=True, ascending=False)
styles['PPDA'] = styles['Press intensity'].rank(pct=True, ascending=False)
styles['Pass out'] = styles['Deep buildup'].rank(pct=True, ascending=False)
styles['Central prog'] = styles['Central progression'].rank(pct=True, ascending=False)
styles['Slow pass'] = styles['Circulate'].rank(pct=True, ascending=False)
styles['Patience'] = styles['Patient attack'].rank(pct=True, ascending=False)
styles.head()
```

```python
styles.columns.values
```

```python
pizza=styles[['Squad','Tight def', 'PPDA','High line','Potency','Patience','Shooting','poss_rank','Resistance rank','Pass out','Tilt rank','Slow pass','Central prog']]
pizza['Chance prevention'] = styles['Tight def']*100
pizza['Pressing'] = styles['PPDA']*100
pizza['Defensive height'] = styles['High line']*100
pizza['Chance creation'] = styles['Potency']*100
pizza['Patient attack'] = styles['Patience']*100
pizza['Shot quality'] = styles['Shooting']*100
pizza['Possession'] = styles['poss_rank']*100
pizza['Press resistance'] = styles['Resistance rank']*100
pizza['Deep build-up'] = styles['Pass out']*100
pizza['Field tilt'] = styles['Tilt rank']*100
pizza['Circulate'] = styles['Slow pass']*100
pizza['Central progression'] = styles['Central prog']*100
pizza=pizza.round()
#final df
final_df=pizza[['Squad','Chance prevention','Pressing','Defensive height','Chance creation','Patient attack','Shot quality','Possession','Press resistance','Deep build-up','Field tilt','Circulate','Central progression']]
final_df.head()
```

```python
#Create parameter and values lists
params = list(final_df.columns)
params = params[1:]
params
```

```python
team = final_df.loc[pizza['Squad']=='Wolves'].reset_index()
team = list(team.loc[0])
print(len(team),print(len(params)))
team = team[2:]
print(len(team),print(len(params)))#
values = team
values
```

```python
# color for the slices and text
slice_colors = ["#1A78CF"] * 3 + ["#FF9300"] * 3 + ["#D70232"] * 3 + ["#00cc00"] * 3
text_colors = ["#000000"] * 12 #+ ["#F2F2F2"] * 6
```

```python
# instantiate PyPizza class
baker = PyPizza(
    params=params,                  # list of parameters
    background_color="#EBEBE9",     # background color
    straight_line_color="#EBEBE9",  # color for straight lines
    straight_line_lw=1,             # linewidth for straight lines
    last_circle_lw=0,               # linewidth of last circle
    other_circle_lw=0,              # linewidth for other circles
    inner_circle_size=10            # size of inner circle
)
```

```python
# plot pizza
fig, ax = baker.make_pizza(
    values,                          # list of values
    figsize=(10, 8.5),                # adjust figsize according to your need
    color_blank_space="same",        # use same color to fill blank space
    slice_colors=slice_colors,       # color for individual slices
    value_colors=text_colors,        # color for the value-text
    value_bck_colors=slice_colors,   # color for the blank spaces
    blank_alpha=0.2,                 # alpha for blank-space colors
    kwargs_slices=dict(
        edgecolor="#000000", zorder=2, linewidth=1
    ),                               # values to be used when plotting slices
    kwargs_params=dict(
        color="000000", fontsize=10,
        va="center", alpha=1
    ),                               # values to be used when adding parameter
    kwargs_values=dict(
        color="#000000", fontsize=10, zorder=3,
        bbox=dict(
            edgecolor="#000000", facecolor="cornflowerblue",
            boxstyle="round,pad=0.2", lw=1
        )
    )                                # values to be used when adding parameter-values
)
# add title
fig.text(
    0.515, 0.975, "Feyenoord playing style", size=14,
    ha="center", color="#000000"
)
# add subtitle
fig.text(
    0.515, 0.953,
    "Percentile rank vs big 5 leagues | Season 2023/24",
    size=12,
    ha="center", color="#000000"
)
```
