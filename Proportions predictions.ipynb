```python
import pandas as pd
```

```python
df = pd.read_html('https://fbref.com/en/comps/9/stats/Premier-League-Stats')
```

```python
for idx,table in enumerate(df):
 print("***************************")
 print(idx)
 print(table)
```

```python
for_prem = pd.DataFrame(df[0])
```

```python
against_prem = pd.DataFrame(df[1])
```

```python
for_prem.drop(columns=["Per 90 Minutes"], inplace=True)
for_prem.columns = for_prem.columns.droplevel()
for_pred = for_prem[['Squad','Gls','npxG','PrgP']]
```

```python
against_prem.drop(columns=["Per 90 Minutes"], inplace=True)
against_prem.columns = against_prem.columns.droplevel()
against_pred = against_prem[['Squad','Gls','npxG','PrgP']]
against_pred = against_pred.rename({'Gls':'GlsA', 'npxG':'npxGA','PrgP':'PrgPA'}, axis=1)
against_pred['Squad'] = against_pred['Squad'].str.replace('vs', '')
```

```python
final_pred = pd.merge(for_pred, against_pred, left_index=True, right_index=True, suffixes=('', '_right'))
final_pred.drop(columns=["Squad_right"], inplace=True)
final_pred
```

```python
final_pred['GD'] = final_pred['Gls']/(final_pred['Gls']+final_pred['GlsA'])
final_pred['npxGD']= final_pred['npxG']/(final_pred['npxG']+final_pred['npxGA'])
final_pred['PrgPD']= final_pred['PrgP']/(final_pred['PrgP']+final_pred['PrgPA'])
final_pred.head()
```

```python
prem = final_pred[['Squad','GD', 'npxGD','PrgPD']].copy()
print(prem)
```

```python
prem.GD=prem.GD.round(2)
prem.npxGD=prem.npxGD.round(2)
prem.PrgPD=prem.PrgPD.round(2)
prem
```
